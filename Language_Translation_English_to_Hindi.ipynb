# Preprocessing the data

Tokenization is the process of converting text into a sequence of tokens or subwords suitable for input to a neural network model.

AutoTokenizer.from_pretrained(): This method loads a pre-trained tokenizer from the Hugging Face model hub. It automatically selects the appropriate tokenizer based on the model_checkpoint provided.

```python
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
tokenizer("Hello, this is a sentence!")
```

```python
tokenizer(["Hello, this is a sentence!", "This is another sentence."])
```

```python
with tokenizer.as_target_tokenizer():
    print(tokenizer(["एक्सेर्साइसर पहुंचनीयता अन्वेषक"]))
```

The tokenizer tokenizes the input text into subwords or tokens suitable for input to the translation model. The output is printed, showing the tokenized representation of the input sentence.

**This step is essential for preparing text data for translation by the model.**

```python
max_input_length = 128
max_target_length = 128

source_lang = "en"
target_lang = "hi"


def preprocess_function(examples):
    inputs = [ex[source_lang] for ex in examples["translation"]]
    targets = [ex[target_lang] for ex in examples["translation"]]
    model_inputs = tokenizer(inputs, max_length=max_input_length, truncation=True)

    # Setup the tokenizer for targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=max_target_length, truncation=True)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
preprocess_function(raw_datasets["train"][:2])
```

```python
tokenized_datasets = raw_datasets.map(preprocess_function, batched=True)
```

 The batched=True parameter specifies that the preprocessing should be applied in a batched manner, improving efficiency by processing multiple examples simultaneously.

```python
model = TFAutoModelForSeq2SeqLM.from_pretrained(model_checkpoint)
```

TFAutoModelForSeq2SeqLM is a class provided by the Hugging Face transformers library for TensorFlow. It represents an auto-compatible sequence-to-sequence model for language modeling tasks.

```python
batch_size = 16
learning_rate = 2e-5
weight_decay = 0.01
num_train_epochs = 1
```

```python
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, return_tensors="tf")
```

```python
generation_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, return_tensors="tf", pad_to_multiple_of=128)
```

```python
train_dataset = model.prepare_tf_dataset(
    tokenized_datasets["test"],
    batch_size=batch_size,
    shuffle=True,
    collate_fn=data_collator,
)
```

```python
validation_dataset = model.prepare_tf_dataset(
    tokenized_datasets["validation"],
    batch_size=batch_size,
    shuffle=False,
    collate_fn=data_collator,
)
```

```python
generation_dataset = model.prepare_tf_dataset(
    tokenized_datasets["validation"],
    batch_size=8,
    shuffle=False,
    collate_fn=generation_data_collator,
)
```

```python
optimizer = AdamWeightDecay(learning_rate=learning_rate, weight_decay_rate=weight_decay)
model.compile(optimizer=optimizer)
```

```python
model.fit(train_dataset, validation_data=validation_dataset, epochs=1)
```

```python
model.save_pretrained("tf_model/")
```
